```
import SimpleITK as sitk 
from tqdm import tqdm
import numpy as np
```

```
def load_data(path):
  my_dir = sorted(os.listdir(path))
  data = []
  gt = []
  for p in tqdm(my_dir):
    data_list = sorted(os.listdir(path+p))
    img_itk = sitk.ReadImage(path + p + '/'+ data_list[0])
    flair = sitk.GetArrayFromImage(img_itk)
    img_itk = sitk.ReadImage(path + p + '/'+ data_list[1])
    seg =  sitk.GetArrayFromImage(img_itk)
    img_itk = sitk.ReadImage(path + p + '/'+ data_list[2])    
    t1 =  sitk.GetArrayFromImage(img_itk)
    img_itk = sitk.ReadImage(path + p + '/'+ data_list[3])
    t1ce =  sitk.GetArrayFromImage(img_itk)
    img_itk = sitk.ReadImage(path + p + '/'+ data_list[4])
    t2 =  sitk.GetArrayFromImage(img_itk)
    data.append([flair,t1,t1ce,t2])
    gt.append(seg)
  data = np.asarray(data,dtype=np.float32)
  gt = np.asarray(gt,dtype=np.uint8)
  return data,gt
```

```
from keras.models import Model
from keras.layers import MaxPooling2D,Conv2D,Dense,BatchNormalization,concatenate,Input,Dropout,Maximum,Activation,Dense,Flatten, Concatenate,UpSampling2D,Conv2DTranspose
from keras.optimizers import SGD,Adam,RMSprop
import keras.callbacks as callbacks
import keras.initializers as initializers
from keras.callbacks import Callback
from keras import regularizers
```

```
from keras import backend as K
def dice_coef(y_true, y_pred, epsilon=1e-6):
    intersection = K.sum(K.abs(y_true * y_pred), axis=-1)
    return (2. * intersection) / (K.sum(K.square(y_true),axis=-1) + K.sum(K.square(y_pred),axis=-1) + epsilon)

def dice_coef_loss(y_true, y_pred):
    return 1-dice_coef(y_true, y_pred)
```

```
def double_conv(x, filters):
    conv = Conv2D(filters, (3, 3), activation='relu', padding='same')(x)
    conv = Conv2D(filters, (3, 3), activation='relu', padding='same')(conv)
    return conv
```

```
inputs = Input((192, 192, 4))
unet_filters = [64, 128, 256, 512, 1024]
    
conv = Conv2D(unet_filters[0], (3, 3), activation='relu', padding='same', input_shape = (192,192, 4))(inputs)
conv1 = Conv2D(unet_filters[0], (3, 3), activation='relu', padding='same')(conv)
norm1 = BatchNormalization()(conv1)
pool1 = MaxPooling2D(pool_size=(2, 2), strides = 2)(norm1)

conv2 = double_conv(pool1, unet_filters[1])
norm2 = BatchNormalization()(conv2)
pool2 = MaxPooling2D(pool_size=(2, 2), strides = 2)(norm2)

dropout1 = Dropout(0.2)(pool2)

conv3 = double_conv(dropout1, unet_filters[2])
norm3 = BatchNormalization()(conv3)
pool3 = MaxPooling2D(pool_size=(2, 2), strides = 2)(norm3)

conv4 = double_conv(pool3, unet_filters[3])
norm4 = BatchNormalization()(conv4)
pool4 = MaxPooling2D(pool_size=(2, 2), strides = 2)(norm4)

conv5 = Conv2D(unet_filters[4], (3, 3), activation='relu', padding='same')(pool4)
dropout2 = Dropout(0.2)(conv5)
#    conv5 = Conv2D(unet_filters[4], (3, 3), activation='relu', padding='same')(pool4)


up_samp = Conv2DTranspose(unet_filters[3], (3,3), strides = 2, padding='same', activation='relu')(dropout2)
concat = Concatenate()([norm4,up_samp])
   
conv6 = double_conv(concat, unet_filters[3])

up_samp2 = Conv2DTranspose(unet_filters[2], (3,3), strides = 2, padding='same', activation='relu')(conv6)
concat2 = Concatenate()([conv3, up_samp2])
conv7 = double_conv(concat2, unet_filters[2])

up_samp3 = Conv2DTranspose(unet_filters[1], (3,3), strides = 2, padding='same', activation='relu')(conv7)
concat3 = Concatenate()([conv2, up_samp3])
conv8 = double_conv(concat3, unet_filters[1])

up_samp4 = Conv2DTranspose(unet_filters[0], (3,3), strides = 2, padding='same', activation='relu')(conv8)
concat4 = Concatenate()([conv1, up_samp4])
conv9 = double_conv(concat4, unet_filters[0])

conv10 = Conv2D(1, (1, 1), activation='sigmoid')(conv9)

    
model = Model(inputs=[inputs], outputs=[conv10])
model.save_weights("weights.hdf5")
```

```
model.compile(optimizer=Adam(lr=1e-5),loss=dice_coef_loss,metrics=[dice_coef])
model.load_weights('weights.hdf5')
from keras.callbacks import ModelCheckpoint
checkpointer = callbacks.ModelCheckpoint(filepath = 'weights.hdf5',save_best_only=True)
training_log = callbacks.TensorBoard(log_dir='./Model_logs')
```

```
history = model.fit(X_train,Y_train,validation_data=(X_val,Y_val),batch_size=32,epochs=16,callbacks=[training_log,checkpointer],shuffle=True)
```

```
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('dice coef loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
plt.plot(history.history['dice_coef'])
plt.plot(history.history['val_dice_coef'])
plt.title('model score')
plt.ylabel('dice coef')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
